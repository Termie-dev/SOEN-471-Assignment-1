```python
#Import modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path ='customer_churn.csv'
data = pd.read_csv(file_path)
#Performing exploratory data analysis (EDA)
# .Display summary statistics
summary = data.describe() # gets a summary of the entire data
print(summary)
```

```python
# .Identifying missing values in the data frame
#Identifying if there is any Null values:
print(data.isnull().any()) # No actual null values found
```

```python
# check for duplicates : 
duplicates = data.duplicated().sum()
print(duplicates) # No duplicates found
```

```python
# .Visualize data distributions using histograms and box plots.
# Check for age distributions:
age_data = data['Age'].tolist()
n, bins, patches  = plt.hist(age_data, bins = 10, edgecolor='black' )
for i in range(len(n)):
    plt.text(bins[i] + 2.2, n[i] + 0.3, str(int(n[i])), ha='center', va='bottom', color ='darkblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show();
```

```python
#Box plot
plt.boxplot(age_data)
plt.title('Age Distribution')
plt.ylabel('Frequency')
plt.show();
```
